```python
from spineq.optimise import optimise, calc_oa_weights, get_optimisation_inputs
from spineq.genetic import build_problem, run_problem, extract_all
from spineq.utils import total_coverage
import pandas as pd
import numpy as np
from tqdm import tqdm
import pygmo as pg
import matplotlib.pyplot as plt
```

```python
n_sensors = 55
theta = 500
pop_age_groups = {"pop_elderly": {"min": 66, "max": 90, "weight": 1}}
```

## Greedy

```python
steps = np.linspace(0, 1, 101)
pop_oa_w = calc_oa_weights(population_weight=1, workplace_weight=0, pop_age_groups=pop_age_groups).sort_index()
work_oa_w = calc_oa_weights(population_weight=0, workplace_weight=1, pop_age_groups=pop_age_groups).sort_index()

pop_cov = []
work_cov = []
pop_weight = []
work_weight = []
for pw in tqdm(steps):
    ww = 1 - pw
    r = optimise(
        n_sensors=n_sensors,
        population_weight=pw,
        workplace_weight=ww,
        theta=theta,
        pop_age_groups=pop_age_groups,
    )
    oa_cov = pd.DataFrame(r["oa_coverage"]).set_index("oa11cd")["coverage"].sort_index()
    pop_weight.append(pw)
    work_weight.append(ww)
    pop_cov.append(total_coverage(oa_cov.sort_index(), pop_oa_w))
    work_cov.append(total_coverage(oa_cov.sort_index(), work_oa_w))
```

```python
plt.plot(pop_cov, work_cov, "ro")
```

## Genetic

```python
inputs = get_optimisation_inputs(
    population_weight=1,
    workplace_weight=1,
    pop_age_groups=pop_age_groups,
    combine=False,
)
prob = build_problem(inputs, n_sensors=n_sensors, theta=theta)
pop = run_problem(
    prob,
    uda=pg.nsga2(gen=7000),
    population_size=200,
    verbosity=1,
)
```

```python
scores, solutions = extract_all(pop)
scores = -scores

plt.plot(scores[:, 0], scores[:, 1], "bo")
```

```python
plt.figure(figsize=(10, 10))
plt.plot(pop_cov, work_cov, "ro", label="greedy")
plt.plot(scores[:, 0], scores[:, 1], "bo", label="genetic")
plt.legend()
plt.xlabel("pop_cov")
plt.ylabel("work_cov")
```

```python
tot_cov_gr = []
tot_cov_ge = []

for pw, pc, ww, wc in zip(pop_weight, pop_cov, work_weight, work_cov):
    tot_cov_gr.append((pw * pc + ww * wc) / (pw + ww))
    
    max_ge = ((scores[:, 0] * pw + scores[:, 1] * ww) / (pw + ww)).max()
    tot_cov_ge.append(max_ge)

plt.plot(pop_weight, tot_cov_gr, "r", label="greedy")
plt.plot(pop_weight, tot_cov_ge, "b", label="genetic")
plt.legend()
plt.xlabel("pop_weight")
```
